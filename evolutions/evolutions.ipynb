```python
import pandas as pd

df = pd.read_csv("../dex/dex.csv", index_col="name")
df = df[["num", "prevo", "evoLevel", "forme", "baseSpecies", "requiredItem", "evoCondition", "evoType", "evoItem", "numEvos", "legendary", "eggGroup_Undiscovered", "canHatch"]]
df = df[df.num > 0]
df
```

```python
df["stage"] = 0
df.loc[~df.prevo.isna(), "stage"] = 1
for name in df[~df.prevo.isna()].index:
    prevo = df.loc[name, "prevo"]
    preprevo = df.loc[prevo, "prevo"]
    if not pd.isna(preprevo):
        df.loc[name, "stage"] = 2
for name in df[~df.forme.isna() & df.forme.str.contains("Mega|Gmax", regex=True)].index:
    baseSpecies = df.loc[name, "baseSpecies"]
    base_stage = df.loc[baseSpecies, "stage"]
    df.loc[name, "stage"] = base_stage + 1

df
```

```python
df.loc[(df.stage==0) & ~df.numEvos.isna(), "evolution_category"] = "Unevolved"
df.loc[(df.stage==1) & ~df.numEvos.isna(), "evolution_category"] = "Middle evolution"
df.loc[(df.stage>=1) & df.numEvos.isna(), "evolution_category"] = "Fully evolved"
df.loc[(df.stage==0) & df.numEvos.isna(), "evolution_category"] = "Does not evolve"
df.loc[(df.stage==0) & df.canHatch, "evolution_category"] = "Baby"
df.loc[~df.forme.isna() & df.forme.str.contains("Mega"), "evolution_category"] = "Mega"
df.loc[~df.forme.isna() & df.forme.str.contains("Gmax"), "evolution_category"] = "Gmax"
df
```

```python
df.loc[~df.evoLevel.isna(), "evolution_method"] = df.evoLevel.apply(lambda x: f"Lv {x:.0f}")
df.loc[df.evoType=="levelFriendship", "evolution_method"] = "Lv up with high happiness"
df.loc[df.evoType=="useItem", "evolution_method"] = df.evoItem
df.loc[(df.evoType=="trade") & df.evoItem.isna(), "evolution_method"] = "Trade"
df.loc[(df.evoType=="trade") & ~df.evoItem.isna(), "evolution_method"] = "Trade while holding " + df.evoItem
df.loc[df.evoType=="levelHold", "evolution_method"] = "Lv up while holding " + df.evoItem
df.loc[df.evoType=="levelMove", "evolution_method"] = "Lv up while having learned "
df.loc[df.evoType=="levelExtra", "evolution_method"] = "Lv up"
df.loc[df.evoType=="other", "evolution_method"] = df.evolution_method

df.loc[~df.evoCondition.isna(), "evolution_method"] = df.evolution_method + " " + df.evoCondition

df.loc[~df.forme.isna() & df.forme.str.contains("Mega"), "evolution_method"] = df.requiredItem
df.loc[~df.forme.isna() & df.forme.str.contains("Gmax"), "evolution_method"] = "Gigantamax Factor"
```

```python
evo_conds = df.evoCondition.value_counts()
evo_conds = evo_conds[evo_conds <= 2]
df.loc[df.evoCondition.isin(evo_conds.index), "special_evolution"] = df.evolution_method
df[~df.special_evolution.isna()]
```

```python
out = pd.read_csv("../template.csv", encoding="utf-8")
out = out[["Species"]]
out["Species"] = out.Species.str.replace("(.*) - (F|M)$", "\g<1>", regex=True)
out["Species"] = out.Species.str.replace("\s+-\s+", "-", regex=True)
out["Species"] = out.Species.str.replace("\s+", "-", regex=True)
out["Species"] = out.Species.str.replace("-Cap", "", regex=True)
out = out.merge(right=df[["evolution_category", "stage", "evolution_method", "special_evolution"]], how="left", left_on="Species", right_index=True)
out.to_csv("evos.csv", index=False)
out[out.evolution_category.isna()]
```
