```python
import pandas as pd

df = pd.read_csv("../dex/dex.csv", index_col="name")
df = df[["num", "prevo", "evoLevel", "forme", "baseSpecies", "requiredItem", "evoCondition", "evoType", "numEvos", "legendary", "eggGroup_Undiscovered", "canHatch"]]
df = df[df.num > 0]
df
```

```python
df["stage"] = 0
df.loc[~df.prevo.isna(), "stage"] = 1
for name in df[~df.prevo.isna()].index:
    prevo = df.loc[name, "prevo"]
    preprevo = df.loc[prevo, "prevo"]
    if not pd.isna(preprevo):
        df.loc[name, "stage"] = 2
# for name in df[~df.forme.isna() & df.forme.str.contains("Mega|Gmax", regex=True)].index:
#     baseSpecies = df.loc[name, "baseSpecies"]
#     base_stage = df.loc[baseSpecies, "stage"]
#     df.loc[name, "stage"] = base_stage + 1
df.loc[~df.forme.isna() & df.forme.str.contains("Mega"), "stage"] = "Mega"
df.loc[~df.forme.isna() & df.forme.str.contains("Gmax"), "stage"] = "Gmax"
df
```

```python
df.loc[(df.stage==0) & df.canHatch, "is_baby"] = "Baby"
df
```
